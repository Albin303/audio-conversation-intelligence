Cell 1 - verify that ffmpeg is available on PATH

```python
import shutil
from pathlib import Path

ffmpeg_path = shutil.which("ffmpeg")
if not ffmpeg_path:
    raise RuntimeError("Install ffmpeg and add it to PATH before running this notebook.")
print("ffmpeg found at:", ffmpeg_path)
```

Cell 2
 load Whisper model

```python
import whisper
model = whisper.load_model("base") 
```

Cell 3 set the audio path (relative to the notebook folder)

```python
audio_path = Path("../audio/conv_001.wav")  # notebooks -> .. -> audio
print("Exists:", audio_path.exists(), "Path:", audio_path)
```

Cell 4  transcribe

```python
result = model.transcribe(str(audio_path))
print(result["text"])
```

Save transcript to the canonical raw-data directory

```python
out_file = Path("../data/raw/conv_001_laptop_sample.txt")
out_file.parent.mkdir(parents=True, exist_ok=True)
out_file.write_text(result["text"], encoding="utf-8")
print("Saved to", out_file)
```
